OpenCV로 동영상 읽기

```python
import cv2
import matplotlib.pyplot as plt

video_path = "/content/drive/MyDrive/시간복잡도를 고려한 openCV 구현/카더가든과 이진이.mp4"

image = cv2.VideoCapture(video_path)

image_name_list = []

while image.isOpened():
  ret, frame = image.read()
  image_name_list.append(frame)
  if ret == False:
    break


print(plt.imshow(image_name_list[-30]))

```
