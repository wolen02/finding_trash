```python
import cv2
import time
import threading
import queue

video_path = "/content/drive/MyDrive/test/video.MP4"
out_path   = "/content/drive/MyDrive/test/gray_video.mp4"

cap = cv2.VideoCapture(video_path)

# 입력 영상 속성 그대로 가져오기 (고정값 하드코딩 금지)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
if fps is None or fps <= 0:
    fps = 30.0  # fallback

# Colab에서 mp4 출력은 보통 mp4v가 가장 무난
fourcc = cv2.VideoWriter_fourcc(*"mp4v")

# 그레이스케일로 저장하려면 isColor=False로 생성
out = cv2.VideoWriter(out_path, fourcc, fps, (w, h), isColor=False)

# 큐(Backpressure)
frame_q = queue.Queue(maxsize=64)
STOP = object()

write_lock = threading.Lock()
time_list = []

def reader():
    """VideoCapture는 한 스레드에서만 read()"""
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_q.put(frame)  # 큐가 차면 여기서 자연스럽게 대기(메모리 폭주 방지)

    # worker들에게 종료 신호 전달
    for _ in range(NUM_WORKERS):
        frame_q.put(STOP)

def worker(worker_id: int):
    """프레임 처리(그레이 변환)"""
    while True:
        item = frame_q.get()
        try:
            if item is STOP:
                return

            t0 = time.perf_counter()
            gray = cv2.cvtColor(item, cv2.COLOR_BGR2GRAY)

            # VideoWriter는 스레드 세이프가 아니므로 Lock으로 보호
            with write_lock:
                out.write(gray)

            dt_ms = (time.perf_counter() - t0) * 1000
            time_list.append(dt_ms)

        finally:
            frame_q.task_done()

NUM_WORKERS = 2  # CPU 코어 상황에 맞춰 2~8로 조정 가능(Colab은 2~4 권장)

t_start_total = time.perf_counter()

t_r = threading.Thread(target=reader, daemon=True)
workers = [threading.Thread(target=worker, args=(i,), daemon=True) for i in range(NUM_WORKERS)]

t_r.start()
for t in workers:
    t.start()

# reader 종료 대기
t_r.join()

# 큐의 모든 작업이 처리될 때까지 대기
frame_q.join()

# worker 종료 대기
for t in workers:
    t.join(timeout=1.0)

# 자원 해제
cap.release()
out.release()

total_sec = time.perf_counter() - t_start_total
print(f"총 처리 시간: {total_sec:.2f} sec")
if time_list:
    print(f"프레임 처리 평균(ms): {sum(time_list)/len(time_list):.3f}, max(ms): {max(time_list):.3f}, count={len(time_list)}")
print(f"저장 경로: {out_path}")
```

```python
import multiprocessing
import threading
import time

def work(p_number, worker_id, count):
  for i in range(count):
    print(f"프로세스 번호:{p_number}, 일꾼번호:{worker_id}, 현재 일 진행 횟수:{i+1}\n")
    time.sleep(1)

def make_thread(p_number):
  t1 = threading.Thread(target=work, args=(p_number, 1, 3,))
  t2 = threading.Thread(target=work, args=(p_number, 2, 3,))

  t1.start()
  t2.start()
  t1.join()
  t2.join()



start = time.time()
p1 = multiprocessing.Process(target=make_thread, args=(1,))
p2 = multiprocessing.Process(target=make_thread, args=(2,))

p1.start()
p2.start()

p1.join()
p2.join()
end = time.time()

print(f"총 소요시간: {end - start}")
```

```python
import multiprocessing
import threading
import time

def work(p_number, worker_id, count):
  for i in range(count):
    print(f"프로세스 번호:{p_number}, 일꾼번호:{worker_id}, 현재 일 진행 횟수:{i+1}\n")
    time.sleep(1)

def make_thread(p_number):
  t1 = threading.Thread(target=work, args=(p_number, 1, 3,))
  t2 = threading.Thread(target=work, args=(p_number, 2, 3,))

  t1.start()
  t2.start()
  t1.join()
  t2.join()



start = time.time()
p1 = multiprocessing.Process(target=make_thread, args=(1,))
p2 = multiprocessing.Process(target=make_thread, args=(2,))
p3 = multiprocessing.Process(target=make_thread, args=(3,))
p4 = multiprocessing.Process(target=make_thread, args=(4,))

p1.start()
p2.start()
p3.start()
p4.start()

p1.join()
p2.join()
p3.join()
p4.join()
end = time.time()

print(f"총 소요시간: {end - start}")
```
